```python
import math
import random
```

```python
LUCRO_POR_UNID = [30, 50, 40] # (Lucro A, Lucro B, Lucro C)
CONSUMO_HORA_POR_UNID = [2, 4, 3] # (Consumo A, Consumo B, Consumo C)
CONSUMO_MOEDAS_POR_UNID = [3, 2, 4] # (Moedas A, Moedas B, Moedas C)

LIM_HORAS = 100
LIM_MOEDAS = 90

ACOES = [-2, -1, 1, 2]
```

```python
def profit(curr_state: list[int], verbose: bool = False) -> int:
    profit = 0
    hours = 0
    coins = 0
    for i, prod_qtd in enumerate(curr_state):
        profit += LUCRO_POR_UNID[i] * prod_qtd
        hours += CONSUMO_HORA_POR_UNID[i] * prod_qtd
        coins += CONSUMO_MOEDAS_POR_UNID[i] * prod_qtd
    
    penalty = 0
    if hours > LIM_HORAS:
        temp = (hours - LIM_HORAS) / LIM_HORAS
        penalty += temp * profit * 5
    if coins > LIM_MOEDAS:
        temp = (coins - LIM_MOEDAS) / LIM_MOEDAS
        penalty += temp * profit * 5
    
    if verbose:
        print(f"Profit: {profit}, Hours: {hours}, Coins: {coins}, Penalty: {penalty}")
    return profit - penalty
```

```python
def random_neighbor(curr_state: list[int]) -> list[int]:
    temp = curr_state.copy()
    prod = random.randint(0, 2)
    action = random.choice(ACOES)
    if temp[prod] + action >= 0:
        temp[prod] += action
    else:
        temp[prod] += abs(action)
    return temp
```

```python
curr_solution = [0, 0, 0] # (Qtd A, Qtd B, Qtd C)
curr_profit = profit(curr_solution)
best_solution = curr_solution

temp = 1000
min_temp = 5
cooling_rate = .95

while temp > min_temp:
    for _ in range(5): # Iteractions per temperature 
        new_solution = random_neighbor(curr_solution)
        new_profit = profit(new_solution)
        if new_profit > curr_profit:
            curr_solution = new_solution
            curr_profit = new_profit
            if new_profit > profit(best_solution):
                best_solution = new_solution
        else:
            delta = new_profit - curr_profit
            prob = math.exp(delta / temp)
            if random.random() <= prob:
                curr_solution = new_solution
                curr_profit = new_profit
    
    temp *= cooling_rate
```

```python
profit(best_solution, verbose=True)
```
